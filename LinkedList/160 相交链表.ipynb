```python
# 如果链表长度一致，那么分别遍历，第一个相等节点即为交点。所以该法目的为：使两链表当前指针到链表尾的长度相同，然后遍历


# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def getIntersectionNode(self, headA, headB):
        """
        :type head1, head1: ListNode
        :rtype: ListNode
        """
        curA = headA
        curB = headB
        while(curA and curB):
            if curA==curB:
                return curA
            curA = curA.next
            curB = curB.next

        if not curA:
            new = headB
            while(curB):
                new = new.next
                curB = curB.next
            while(new and headA):
                if new==headA:
                    return new
                new = new.next
                headA = headA.next
            return None
        else:
            new = headA
            while(curA):
                new = new.next
                curA = curA.next
            while(new and headB):
                if new==headB:
                    return new
                new = new.next
                headB = headB.next
            return None
        
        
```
