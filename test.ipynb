```python
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        l = len(nums)
        left = 0
        right = l-1
        while(left<=right):

            mid = (left+right)//2
            if target==nums[mid]:
                return mid
            elif target<nums[mid]:
                right = mid-1
            else:
                left = mid+1
        return left
```

```python
class Solution(object):
    def kthSmallest(self, matrix, k):
        """
        :type matrix: List[List[int]]
        :type k: int
        :rtype: int
        """
        n = len(matrix)
        l = matrix[0][0]
        r = matrix[n-1][n-1]
        mid = (l+r)//2
        while(True):
            res = self.calc(matrix,mid,n)
#             print(res)
            if(res==k):
                return mid
            elif(res<k):
                l = mid+1
            else:
                r = mid-1
    def calc(self,matrix,target,n):
        res = 0
        for i in range(n):
            l = 0
            r = n-1
            tmp = 0
            while(l<=r):
                mid = l + (r-l)//2      
                print(l,r,mid)
                if(matrix[i][mid]<target):
                    tmp = mid+1
                    l = mid+1
                else:
                    r = mid-1
            
            res +=tmp
            print("After ",i," row: ",res)
        return tmp
```

```python
class Solution:
    def guess(self, matrix, g, n, k):
        sum = 0
        for i in range(n):
            L = 0
            R = n - 1
            ans = 0
            while (L <= R):
                mid = L + (R - L)//2
                
                if(matrix[i][mid] < g):
                    ans = mid + 1
                    L = mid + 1
                else:
                    R = mid - 1
            sum += ans
        return k > sum;

    def kthSmallest(self,matrix, k):
        n = len(matrix)
        L = matrix[0][0]
        R = matrix[n - 1][n - 1]
        ans = 0
        while (L <= R):
            mid = L + (R - L )/2
            if(self.guess(matrix, mid, n, k)):
                ans = mid
                L = mid + 1
            else:
                R = mid - 1
        return  ans
```
