```python
class TrieNode(object):
    def __init__(self,val=""):
        self.val = val
        self.child = [None]*26
        self.isEnd = False
        self.num = 0
class Trie(object):
    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.root = TrieNode()

    def insert(self, word):
        """
        Inserts a word into the trie.
        :type word: str
        :rtype: None
        """
        curr = self.root
        for w in word:
            idx = ord(w)-ord('a')
            if not curr.child[idx]:
                curr.child[idx] = TrieNode(w)
            curr = curr.child[idx]
            curr.num += 1
        curr.isEnd = True
        

    def search(self, word):
        """
        Returns if the word is in the trie.
        :type word: str
        :rtype: bool
        """
        curr = self.root
        for w in word:
            idx = ord(w)-ord('a')
            if not curr.child[idx]:
                return False
            curr = curr.child[idx]
        return curr.isEnd
        

    def startsWith(self, prefix):
        """
        Returns if there is any word in the trie that starts with the given prefix.
        :type prefix: str
        :rtype: bool
        """
        curr = self.root
        for w in prefix:
            idx = ord(w)-ord('a')
            if not curr.child[idx]:
                return False
            curr = curr.child[idx]
        return True
        


# Your Trie object will be instantiated and called as such:
# obj = Trie()
# obj.insert(word)
# param_2 = obj.search(word)
# param_3 = obj.startsWith(prefix)
```
