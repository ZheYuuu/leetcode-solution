```python
class Solution:
    def maxProfit(self, prices, fee):
        """
        :type prices: List[int]
        :type fee: int
        :rtype: int
        """
        last_buy = prices[0]
        res = 0
        for price in prices:
            if price < last_buy:
                last_buy = price
            else:
                profit = price - last_buy - fee
                if profit > 0:
                    res += profit
#这句才是整个程序的核心，记录上次价格减去服务费后的资产和利益，然后在下次购买前
#减去上次购买的资产和利益看看是否有盈余。每次都和前面比较是否达到了最大利益化，
#配合第一句判断，真是艺术。
                    last_buy = price - fee
        return res
```
