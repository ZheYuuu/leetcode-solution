```python
第 i 个人的体重为 people[i]，每艘船可以承载的最大重量为 limit。

每艘船最多可同时载两人，但条件是这些人的重量之和最多为 limit。

返回载到每一个人所需的最小船数。(保证每个人都能被船载)。

 

示例 1：

输入：people = [1,2], limit = 3
输出：1
解释：1 艘船载 (1, 2)
示例 2：

输入：people = [3,2,2,1], limit = 3
输出：3
解释：3 艘船分别载 (1, 2), (2) 和 (3)
示例 3：

输入：people = [3,5,3,4], limit = 5
输出：4
解释：4 艘船分别载 (3), (3), (4), (5)
提示：

1 <= people.length <= 50000
1 <= people[i] <= limit <= 30000
```

```python
class Solution:
    def numRescueBoats(self, people, limit) -> int:
        people.sort(reverse = True)
        left = collections.Counter(people)
        ret = 0
        for i,p in enumerate(people):
            if not left[p]:
                continue
            curr = limit-p
            left[p] -= 1
            for key in left.keys():
                if p<key:
                    continue
                if curr - key>=0 and left[key]:
                    left[key] -= 1
                    break
            ret += 1
        return ret
                    
                    
        
```

```python
# 想岔了
class Solution:
    def numRescueBoats(self, people, limit) -> int:
        people.sort(reverse = True)
        left = collections.Counter(people)
        ret = 0
        for p in left.keys():
            if not left[p]:
                continue
            route = [p]
            curr = limit-p
            for key in left.keys():
                if p<key:
                    continue
                if curr - key>=0 and left[key]:
                    route.append(key)
                    break
            if len(route)==1:
                ret += left[route[0]]
                left[route[0]] = 0
            else:
                if route[0]!=route[1]:
                    t = min(left[route[0]],left[route[1]])
                    ret += t
                    left[route[0]] -= t
                    left[route[1]] -= t
                else:
                    ret += left[route[0]]//2
                    left[route[0]] = left[route[0]]&1
        return ret
```

```python
class Solution:
    def numRescueBoats(self, people, limit) -> int:
        people.sort(reverse=True)
        l = 0
        r = len(people)-1
        cnt = 0
        while(l<=r):
            curr = limit-people[l]
            if curr>=people[r]:
                r-=1
            l+=1
            cnt+=1
        return cnt
```
